## 📌 Project Overview

The **Indian Premier League (IPL)** is a professional Twenty20 (T20) cricket league in India featuring franchise teams representing different cities. This project explores the **IPL 2022 match-level dataset** to derive meaningful insights into match outcomes, player performances, team strategies, and overall tournament trends.

---

## 🎯 Project Objective

The primary objective of this project is to perform **Exploratory Data Analysis (EDA)** on the IPL 2022 dataset to answer key questions such as:

- Which team won the most matches?
- Which players delivered outstanding performances?
- How did the toss influence match results?
- Which venues witnessed high-scoring games?
- What were the common winning margins?

---

The following are the important columns used for analysis:

| **Column Name** | **Data Type** | **Description** |

| `top_scorer` | String | Highest run scorer in the match |

### Loading the Libraries and Dataset

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore') 

df = pd.read_csv('IPL.csv')
```

### **Now, Here comes basic questions**

#### 1. Which team won the most matches?

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title('Most match win by team')
plt.xlabel('No of Matches')
plt.ylabel('Team') 
plt.show()
```

#### 2. Toss Decision Trends

```python
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title('Toss Decision Trends')
plt.xlabel('Toss Decision') 
plt.ylabel('Count') 
plt.show()
```

#### 3. Toss Winner vs Match Winner

```python
count = df[df['toss_winner']== df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

```python
sns.countplot(x=df['won_by'])
plt.title('Won By')
```

## Key Player Performance

#### 1. Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x=count.values,y=count.index,palette='mako')
plt.title('Top 10 Playes with Man of the Match')
```

#### 2. Two Top Scorers 

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
```

```python
high.plot(kind='bar')
```

#### 10 Best Bowling Figure 

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets']  = df['highest_wickets'].astype(int) 
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)


top_bowlers.plot(kind='bar')
plt.title('10 Best Bowling Figure') 
plt.xlabel('Bowlers') 
plt.ylabel('Wickets Taken') 
```

#### Q3.Which venue hosted the most matches?

```python
plt.figure(figsize=(10,5))
df['venue'].value_counts().plot(kind='bar')
plt.title('Matches Played at Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

#### Q4. Which team won the most matches while batting first?

```python
batting_first = df[df['won_by'] == 'Runs']

batting_first['match_winner'].value_counts().plot(kind='bar', figsize=(8,5))
plt.title('Wins While Batting First')
plt.ylabel('Wins')
plt.show()
```

#### Q5. Which team won the most matches while batting first?
